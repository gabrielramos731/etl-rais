```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
colunas = ['CBO Ocupação 2002', 'CNAE 2.0 Classe', 'Faixa Etária', 'Escolaridade após 2005', 'Qtd Hora Contr', 'Idade', 'Mun Trab', 'Município','Raça Cor', 'Vl Remun Média Nom','Vl Remun Média (SM)','Sexo Trabalhador','Tempo Emprego']
dfs = []
for ano in range(2014, 2017):
    df = pd.read_csv("D:/dados/rais/vinculos/ES" + str(ano) + ".txt", sep=";", encoding="latin1", usecols=colunas)
    
    ocup = ['2123','2122', '1236', '3171', '3172']
    df['CBO Ocupação 2002'] = df['CBO Ocupação 2002'].apply(lambda x: str(x)[:4])
    df = df[df['CBO Ocupação 2002'].isin(ocup)]

    df['Vl Remun Média Nom'] = df['Vl Remun Média Nom'].apply(lambda x: float(x.replace(',', '.')))
    df['Vl Remun Média (SM)'] = df['Vl Remun Média (SM)'].apply(lambda x: float(x.replace(',', '.')))
    df['Tempo Emprego'] = df['Tempo Emprego'].apply(lambda x: float(x.replace(',', '.')))
    df['ano'] = ano
    dfs.append(df)
df = pd.concat(dfs, ignore_index=True)
```

```python
df
```

```python
df.describe()
```

```python
df_idade_maior = df.query('Idade >= 18')
df_idade_maior.groupby('ano').describe()['Vl Remun Média Nom']['mean'].plot(marker='o', 
                                                                 title='Relação entre a média salarial e o ano', 
                                                                 xlabel='Ano', 
                                                                 ylabel='Média Salarial', 
                                                                 grid=True)
plt.xticks(df_idade_maior['ano'].unique())
```

```python
df_idade_menor = df.query('Idade < 18')
df_idade_menor.groupby('ano').describe()['Vl Remun Média Nom']['mean'].plot(marker='o', 
                                                                 title='Relação entre a média salarial e o ano', 
                                                                 xlabel='Ano', 
                                                                 ylabel='Média Salarial', 
                                                                 grid=True)
plt.xticks(df_idade_menor['ano'].unique())
```
